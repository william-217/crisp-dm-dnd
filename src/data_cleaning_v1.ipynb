```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Pré eliminação de colunas não necessárias
df = df_origin.drop(columns=['Unnamed: 0', 
                             'char_id',
                             'name',
                             'class_starting_level',
                             'subclass_starting',
                             'class_other',
                             'subclass_other',
                             'total_level',
                             'feats',
                             'date_modified',
                             'notes_len',
                             'inventory',
                             'gold',
                             'background',])

# Explicação:
# - 'Unnamed: 0', 'char_id', 'name': identificadores e nomes sem utilidade preditiva;
# - 'class_starting_level', 'subclass_starting', 'class_other', 'subclass_other': detalham variações da classe, mas não são o objectivo deste projecto (pretende-se prever apenas a classe principal).
#     Além disso, uma subclasse é sempre dependente de uma classe principal, sendo definida apenas após esta, o que poderia introduzir fuga de informação (data leakage) se usada como preditor;
# - 'total_level': redundante com 'class_starting_level', e sem valor isolado para previsão;
# - 'feats', 'notes_len', 'gold': atributos textuais ou numéricos que resultam de escolhas feitas após a definição da classe, podendo reflectir decisões personalizadas ou evolução posterior da personagem;
# - 'inventory': embora contenha itens que por vezes estão correlacionados com a classe (ex.: spellbook para wizards), o inventário é atribuído em função da classe, pelo que o seu uso como variável explicativa configuraria um caso clássico de fuga de dados (data leakage);
# - 'date_modified': informação temporal irrelevante para o objectivo de classificação;
# Todas estas colunas poderiam introduzir ruído, redundância ou data leakage, comprometendo a integridade e validade do modelo preditivo.
# A coluna 'background' foi removida nesta fase, pois:
# - Pode conter dependências narrativas ou estruturais com a classe (ex.: backgrounds típicos para certas classes),
#   o que introduziria risco de data leakage;
# - Apresenta elevado número de categorias distintas, dificultando o pré-processamento e aumentando o risco de overfitting;
# - Não é essencial para o objetivo principal, que é prever a classe com base apenas nos atributos e na raça.

```

```python
#Lista de classes oficiais
classes_oficiais = [
    "Barbarian", "Bard", "Cleric", "Druid", "Fighter", "Monk",
    "Paladin", "Ranger", "Rogue", "Sorcerer", "Warlock", "Wizard", "Artificer"
]
```

```python
print(df['class_starting'].unique())
```

```python

# Valores únicos no dataset
classes_dataset = set(df['class_starting'].unique())

# Quais classes oficiais estão no dataset?
presentes = [c for c in classes_oficiais if c in classes_dataset]
faltam = [c for c in classes_oficiais if c not in classes_dataset]

print("Classes oficiais presentes no dataset:", presentes)
print("Classes oficiais em falta no dataset:", faltam)
```

```python
# Agrupar Artificer e Blood Hunter
df['class_starting'] = df['class_starting'].replace({
    'Artificer (UA)': 'Artificer',
    'Blood Hunter (archived)': 'Blood Hunter',

})
```

```python
print(df.nunique())
print(df['class_starting'].unique())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
ax = df['class_starting'].value_counts().plot(kind='bar')
plt.title('Distribuição das Classes')
plt.xlabel('Classe')
plt.ylabel('Número de Personagens')
plt.xticks(rotation=45)
plt.tight_layout()

# Adiciona o número exato em cima de cada barra
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

```python

# Tabela de classes e número de personagens:
tabela_classes = df['class_starting'].value_counts().reset_index()
tabela_classes.columns = ['Classe', 'Número de Personagens']
print(tabela_classes)
# NOTA IMPRTANTE: usar class_weight='balanced' no modelo para lidar com o desbalanceamento de classes!
```

Mudificar header para ficar legivél

```python
df = df.rename(columns={
    'stats_1': 'strength',
    'stats_2': 'dexterity',
    'stats_3': 'constitution',
    'stats_4': 'intelligence',
    'stats_5': 'wisdom',
    'stats_6': 'charisma',
    'race': 'raca',
    'class_starting': 'class',
})
print(len(df))
```

Conversão dos dataTypes

```python
df['class'] = df['class'].astype('category')
df['raca'] = df['raca'].astype('category')


```

```python
colunas_numericas = ['base_hp', 'strength', 'dexterity', 'constitution', 'intelligence', 'wisdom', 'charisma']
df[colunas_numericas] = df[colunas_numericas].apply(pd.to_numeric)

# Conversão de tipos de dados:
# - As colunas 'classe' e 'raca' foram convertidas para o tipo 'category', o que reduz o consumo de memória
#   e facilita a codificação em etapas posteriores;
# - As colunas numéricas foram forçadas para tipo numérico explícito com pd.to_numeric, garantindo compatibilidade
#   com operações de normalização e treino dos modelos.
```
